```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import requests

import time
from selenium.common.exceptions import ElementNotInteractableException
from selenium.common.exceptions import  NoSuchElementException
import re
import sys


driver=webdriver.Chrome(r"C:\Users\sayan\OneDrive\Desktop\chromedriver_win32\chromedriver.exe")
driver.get("https://github.com/")
time.sleep(3)

open_source=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button")
open_source.click()

trending_rep=driver.find_element(By.XPATH,'//a[@href="/trending"]')
trending_rep.click()

time.sleep(3)

rep_list=driver.find_elements(By.XPATH,'//h1[@class="h3 lh-condensed"]')

title_list=[]
desc_list=[]
contributers_count_list=[]
language_list=[]

for i in rep_list:
    href=i.find_element(By.TAG_NAME,"a").get_attribute("href")
    #print(href)
    driver.execute_script('''window.open("'''+href+'''","_blank");''')
    driver.switch_to.window(driver.window_handles[1])
    #i.find_element(By.TAG_NAME,"a").click()
    time.sleep(5)
    title=driver.find_element(By.XPATH,"//div[@class=' d-flex flex-wrap flex-items-center wb-break-word f3 text-normal']").text
    title_list.append(" ".join(title.replace("Public","").split("\n")))
    try:
        desc=driver.find_element(By.XPATH,"//p[@class='f4 my-3']").text
        desc_list.append(desc)
    except NoSuchElementException as e:
        desc_list.append("NULL")
        
    
    try:
        contributers_count=driver.find_element(By.XPATH,"/html/body/div[1]/div[4]/div/main/turbo-frame/div/div/div/div[3]/div[2]/div/div[5]/div/h2/a/span").text
        contributers_count_list.append(contributers_count)
    except NoSuchElementException as e:
        contributers_count_list.append("NULL")
        
    
    languages=driver.find_elements(By.XPATH,"//span[@class='color-fg-default text-bold mr-1']")
    strLang=""
    count=0
    for i in languages:
        if count==0:
            strLang=strLang+i.text
            count=count+1
        else:
            strLang=strLang+","+i.text
    language_list.append(strLang)
        

    
    #break
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    
    
df=pd.DataFrame({"Repository title":title_list,"Repository description":desc_list,
                 "Contributors count":contributers_count_list,"Language used":language_list})

df
    
    




```

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import requests

import time
from selenium.common.exceptions import ElementNotInteractableException
from selenium.common.exceptions import  NoSuchElementException
import re
import sys


driver=webdriver.Chrome(r"C:\Users\sayan\OneDrive\Desktop\chromedriver_win32\chromedriver.exe")
driver.get("https:/www.billboard.com/")
time.sleep(12)

button=driver.find_element(By.XPATH,"//a[@href='https://www.billboard.com/charts/billboard-200/']")
driver.execute_script("arguments[0].click();", button)

song_name_list=[]
artist_list=[]
last_week_rank_list=[]
peak_rank_list=[]
weeks_on_board_list=[]

try:
    songs=driver.find_elements(By.XPATH,"//h3[@class='c-title  a-no-trucate a-font-primary-bold-s u-letter-spacing-0021 lrv-u-font-size-18@tablet lrv-u-font-size-16 u-line-height-125 u-line-height-normal@mobile-max a-truncate-ellipsis u-max-width-330 u-max-width-230@tablet-only']")
except NoSuchElementException as e:
    print("song exception")
for i in songs[0:100]:
    song_name_list.append(i.text)
    
try:
    artists=driver.find_elements(By.XPATH,'//span[@class="c-label  a-no-trucate a-font-primary-s lrv-u-font-size-14@mobile-max u-line-height-normal@mobile-max u-letter-spacing-0021 lrv-u-display-block a-truncate-ellipsis-2line u-max-width-330 u-max-width-230@tablet-only"]')
except NoSuchElementException as e:
    print("artist exception")
for i in artists[0:100]:
    artist_list.append(i.text)

try:
    row_data=driver.find_elements(By.XPATH,'//span[@class="c-label  a-no-trucate a-font-primary-s lrv-u-font-size-14@mobile-max u-line-height-normal@mobile-max u-letter-spacing-0021 lrv-u-display-block a-truncate-ellipsis-2line u-max-width-330 u-max-width-230@tablet-only"]' )
except NoSuchElementException as e:
    print("rank exception")
    
for i in row_data[0:100]:
    parent=i.find_element(By.XPATH,"..")
    grandparent=parent.find_element(By.XPATH,"..")
    li_s=grandparent.find_elements(By.TAG_NAME,"li")
    count=0
    for i in li_s:
        
        if count==3:
            last_week_rank_list.append(i.text)
            
        if count==4:
            peak_rank_list.append(i.text)
            
        if count==5:
            weeks_on_board_list.append(i.text)
        
        
        count=count+1
    
df=pd.DataFrame({"Song name":song_name_list[0:100],"Artist name":artist_list[0:100],"Last week rank":last_week_rank_list[0:100],"Peak rank":peak_rank_list[0:100],"Weeks on board":weeks_on_board_list[0:100]})
df

```
