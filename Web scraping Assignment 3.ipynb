```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import requests

import time
from selenium.common.exceptions import ElementNotInteractableException
from selenium.common.exceptions import  NoSuchElementException

```

```python
#Project1 Project2
driver=webdriver.Chrome(r"C:\Users\sayan\OneDrive\Desktop\chromedriver_win32\chromedriver.exe")
driver.get("https://www.amazon.in/")

#putting search value
search=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")

a=input("Please enter a item to search: \n")

search.send_keys(a)
time.sleep(2)
#clicking search value
search_button=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input")
search_button.click()
time.sleep(3)


brandTags=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')

index=0

brandList=[]
productNameList=[]
priceList=[]
returnExchangeList=[]
expectedDeliveryList=[]
availabilityList=[]
productUrlList=[]
delay=10

try:
    next=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[1]/div[1]/div/span[1]/div[1]/div[31]/div/div/span/a[3]')     
except NoSuchElementException as e:
    next=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')

for i in range(1,3):
    for i in brandTags:
        try:
            
            price=driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')
            priceList.append(price.text)
    
            expectedDelivery=driver.find_element(By.XPATH,'//span[@class="a-color-base a-text-bold"]')
            expectedDeliveryList.append(expectedDelivery.text)
            productUrl=driver.find_element(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]').get_attribute("href") 
            #print(productUrl)
            productUrlList.append(productUrl)
            i.click()
            print("i===================== ",index)
            index=index+1
        
           #transfer control over active window
            driver.switch_to.window(driver.window_handles[index])
            time.sleep(2)
            productName=driver.find_element(By.XPATH,'//span[@id="productTitle"]')
           # print("productName---",productName.text)
            brandList.append(productName.text.split(" ")[0])
            productNameList.append(productName.text)

            returnExchange=driver.find_element(By.XPATH,'//a[@class="a-size-small a-link-normal a-text-normal"]')
            returnExchangeList.append(returnExchange.text)

            availability=driver.find_element(By.XPATH,'//span[@class="a-size-medium a-color-success"]')
            availabilityList.append(availability.text)
            driver.switch_to.window(driver.window_handles[0])
          #  print("------------------------")
    
        except ElementNotInteractableException as e:
            
          #  print("exception occured")
    
            driver.find_element(By.XPATH,'//a[@class="a-link-normal s-no-outline"]').click()
        
        except NoSuchElementException as e:
         #   print("exception occured",e)
            continue
        except StaleElementReferenceException as e:
         #   print("exception occured",e)
            continue
        

    next.click()
    time.sleep(5)


   
    
print(brandList,'\n',productNameList,'\n',priceList,'\n',returnExchangeList,'\n',expectedDeliveryList,'\n',availabilityList)

df=pd.DataFrame({'Brand':brandList,'Product Name':productNameList,'Price':priceList,'Return/Exchange':returnExchangeList,
                'Expected Delivery':expectedDeliveryList,'Availability':availabilityList,'Product URL':productUrlList})
df
```

```python
#Project3
import urllib
from selenium.webdriver.common.keys import Keys
import os

driver=webdriver.Chrome(r"C:\Users\sayan\OneDrive\Desktop\chromedriver_win32\chromedriver.exe")


#creating a directory to save images
directory = "sayan_img1"
# Parent Directory path
parent_dir = "Documents"
# Path
path = os.path.join(parent_dir, directory)

try: 
    os.mkdir(path) 
except OSError as error: 
    print(error)
    
print("Directory '% s' created" % directory)



def download_image(url, folder_name, num):
    urllib.request.urlretrieve(url, path+"/"+"My_image_"+str(num)+".jpg")


def input_search_keyword(key):
    driver.get("https://www.google.com/")
    search=driver.find_element(By.XPATH,"/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input")

    search.send_keys(key)
    search.send_keys(Keys.RETURN);
    time.sleep(2)

    try:
        images=driver.find_element(By.XPATH,"//a[@class='zDN4ab JNHcG']");
    except NoSuchElementException as e:
        images=driver.find_element(By.XPATH,"/html/body/div[7]/div/div[4]/div/div[1]/div/div[1]/div/div[2]/a");

    images.click()
    
    for i in range(1,10):
        elements=driver.find_elements(By.XPATH,"//img[@class='rg_i Q4LuWd']")
        im=elements[i]
        try:
            download_image(im.get_attribute("src"), directory, i)
            print("Downloaded ")
        except e:
            print("exception occured",e)
            print("Couldn't download an image %s, continuing downloading the next one"%(i))
        

searchList=['fruits','cars','Machine Learning','Cakes']

    
for i in searchList:
    input_search_keyword(i)
    
    
        



```
