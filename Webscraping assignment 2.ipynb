```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
#Project1
driver=webdriver.Chrome(r"C:\Users\sayan\OneDrive\Desktop\chromedriver_win32\chromedriver.exe")
driver.get("https://www.naukri.com/")
designation=driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation.send_keys('Data Analyst')

location=driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")
location.send_keys('Bangalore')

search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
time.sleep(4)

jobTitle=[]

titleTags=driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')

for i in titleTags[0:10]:
    title=i.text
    jobTitle.append(title)
    
companyTags=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')

companyList=[]
for i in companyTags[0:10]:
    company=i.text
    companyList.append(company)
    
locationTags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft fs12 lh16 locWdth"]')
locationList=[]
for i in locationTags[0:10]:
    location=i.text
    locationList.append(location)
    
experienceTags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft fs12 lh16 expwdth"]')

experienceList=[]
for i in experienceTags[0:10]:
    experience=i.text
    experienceList.append(experience)
    

df=pd.DataFrame({'Job Title':jobTitle,'Campany':companyList,'Year of experience':experienceList,'Location':locationList})
df
```
